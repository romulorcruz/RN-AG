### Criando classes:

Criando a classe 'Element', na qual será necessário declarar o seu símbolo, número atômico e massa atômca. Também criei uma classe 'Molecule' que recebe um dicionário com o nome do elemento e suas propriedades.

```python
# Criando a classe Element conforme o enunciado e armazendando os parêmetros inseridos como instâncias
class Element:
    def __init__(self,symbol:str,a_number:int,a_mass:float):
        self.symbol = symbol
        self.a_number = a_number
        self.a_mass = a_mass
```

```python
# Criando a classe Molecule baseado no enunciado (com entrada um dicionário, que será armazenado na instância amount)
class Molecule:
    def __init__(self, dictionary:dict):
        self.amount = dictionary 
                
    def show_mass(self): # Função que mostra a massa molecular
        self.mass = 0
        for element in self.amount:
            self.mass += self.amount.get(element) * element.a_mass
        return self.mass

    def generate_formula(self): # Função de gerar fórmula, que pegará a chave(nome do elemento) e o seu valor respectivo(quantidade de átomos)
        self.formula = ''
        for element in self.amount:
            self.formula += element.symbol
            if self.amount.get(element) > 1:
                self.formula += str(self.amount.get(element))
        return self.formula
```

### Criando objetos utilizando as classes criadas:

Coloquei um 'for' para criar um objeto da classe 'Element' para cada elemento químico presente no 'DataFrame' da tabela periódica, iterando suas respectivas informações

```python
# Criando um objeto Element para cada elemento da tabela periódica com base nos dados da tabela periódica
for line in range(0,len(p_table)):
    globals()[p_table['Name'][line]] = Element(p_table['Symbol'][line], p_table['AtomicNumber'][line], p_table['AtomicMass'][line])
```

Como a classe 'Molecule' pega um dicionário como parâmetro para criar o objeto molécula, será criado um dicionário para cada molécula presente no 'DataFrame' de moléculas. Depois disso, será criado um objeto de classe 'Molecule' para cada molécula presente no 'DataFrame' de moléculas utilizando o próprio dicionario homônimo obtido anteriormente, para que assim sobrescreva a variável, reduzindo o armazenamento desnecessário desses dicionários.

```python
for molecule in range(len(molecules)):
    globals()[molecules.iloc[molecule, 0]] = {} # Cria dicionário vazio
    for column in range(1, len(molecules.columns)): # Estrutura de repetição que preencherá os dicionários
        if molecules.iloc[molecule, column] > 0: # Condição para preencher somente átomos presentes na molécula
            column_name =  globals().get(molecules.columns[column]) # Itera o nome do elemento e cria a chave com esse nome
            globals()[molecules.iloc[molecule, 0] ][column_name] = molecules.iloc[molecule, column].astype(int) # Itera o valor da quantidade de átomos presentes na molécula e dá o valor da chave
    globals()[molecules.iloc[molecule, 0]] = (Molecule(globals()[molecules.iloc[molecule, 0]])) # Sobreescreve o dicionário com um objeto 'Molecule'
```

Esse objeto será criado com o nome da molécula que está armazendado na coluna [0], a sua linha será representada por 'molecule', um inteiro que representa a linha em que este elemento está presente.
